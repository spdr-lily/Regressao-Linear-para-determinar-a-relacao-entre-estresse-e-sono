# Regressão Linear
O obejetivo desse projeto é modelar a relação entre estresse e sono, com dados simulados, usando regressão linear simples.

```python
# importação de bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
# variavel independente: estresse (0 a 10)
X = np.random.randint(0, 10, 100)
```

```python
# relação linear: mais estresse -> menos sono
beta_0 = 8 # intercepto: sono médio com estresse 0
beta_1 = 0.5 # cada ponto de estresse reduz 0.5h de sono
```

```python
# erro aleatório
erro = np.random.normal(0, 1, 100)
```

```python
# variável dependente: horas de sono
Y = beta_0 + beta_1 * X + erro
```

```python
# dataset
df = pd.DataFrame({'estresse': X, 'sono': Y})
```

```python
# visualização dos dados
sns.set(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='estresse', y='sono', color='blue')
plt.title('Relação entre Estresse e Sono')
plt.xlabel('NívelEstresse')
plt.ylabel('Horas de Sono')
plt.show()
```

```python
# aplicação da regressão linear
from sklearn.linear_model import LinearRegression

# ajuste do modelo
X_reshape = X.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(X_reshape, Y)
```

```python
# coeficientes
print('Intercepto (beta_0):', modelo.intercept_)
print('Coeficiente angular (beta_1):', modelo.coef_[0])
```

```python
# plot da reta de regressão
Y_pred = modelo.predict(X_reshape)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='estresse', y='sono', color='blue', label='Dados Observados')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Reta de Regressão')
plt.title('Relação Liear: Estresse vs Sono')
plt.xlabel('Nível de Estresse')
plt.ylabel('Horas de Sono')
plt.legend()
plt.show()
```

```python
# métricas de avaliação
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y, Y_pred)
r2 = r2_score(Y, Y_pred)

print('Erro Quadrático Médio (MSE):', mse)
print('Coeficiente de Determinação (R^2):', r2)
```
